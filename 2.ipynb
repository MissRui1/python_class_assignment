```python
import pandas as pd
import numpy as np

from  IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
%%time
data = pd.read_csv('D:/python_class_2025/000001.csv')
data["Day"] = pd.to_datetime(data["Day"],format="%Y/%m/%d")
data.set_index("Day",inplace=True)
data.sort_index(ascending=True,inplace=True)
data
```

```python
data_new = data['1997-01':'2023-12'].copy()
data_new['Close'] = pd.to_numeric(data_new['Close'],errors='coerce')
data_new['Preclose'] = pd.to_numeric(data_new['Preclose'],errors='coerce')
data_new
```

日收益率

```python
data_new['Raw_return'] = data_new['Close']/data_new['Preclose'] - 1
data_new['Log_return'] = np.log(data_new['Close']/data_new['Preclose'])
data_new['Percent_return'] = data_new['Close'].pct_change()
data_new
```

月度收益率

```python
Month_data = data_new.resample('ME')['Log_return'].sum().to_frame()
Month_data.rename(columns={'Log_return':'Month_Log_return'},inplace=True)
Month_data
```

```python
Month_data2 = data_new.resample('ME')['Close'].last().to_frame()
Month_data2['Preclose'] = Month_data2['Close'].shift(1)
Month_data2['Month_Raw_return'] = Month_data2['Close']/Month_data2['Preclose'] - 1
Month_data2
```

group by 方法

```python
# data_new = data['1997-01':'2023-12'].copy()
# data_new['year'] = data_new.index.year
# data_new['month'] = data_new.index.month
# data_new['Log_return'] = np.log(data_new['Close']/data_new['Preclose'])     
# data_new['Raw_return'] = data_new['Close']/data_new['Preclose'] - 1
# data_new

data_new = data['1997-01':'2023-12'].copy()

# 确保 'Close' 和 'Preclose' 列为数值类型
data_new['Close'] = pd.to_numeric(data_new['Close'], errors='coerce')
data_new['Preclose'] = pd.to_numeric(data_new['Preclose'], errors='coerce')

# 删除包含空值的行
data_new.dropna(subset=['Close', 'Preclose'], inplace=True)

# 添加年份和月份列
data_new['year'] = data_new.index.year
data_new['month'] = data_new.index.month

# 计算对数收益率和原始收益率
data_new['Log_return'] = np.log(data_new['Close'] / data_new['Preclose'])
data_new['Raw_return'] = data_new['Close'] / data_new['Preclose'] - 1

data_new
```

```python
Month_data2 = data_new.groupby(['year', 'month'])['Log_return'].sum().to_frame()
Month_data2.rename(columns={'Log_return': 'Month_Log_return'}, inplace=True)    
Month_data2['Raw_return'] = np.exp(Month_data2['Month_Log_return']) - 1
Month_data2
```

季度收益率

```python
Quarter_data = data_new.resample('QE').agg({'Close': 'last', 'Log_return': 'sum'})
Quarter_data
```

```python
Quarter_data = data_new.resample('QE').agg({'Close': 'last', 'Log_return': 'sum'})
Quarter_data['Preclose'] = Quarter_data['Close'].shift(1)
Quarter_data['Raw_return'] = Quarter_data['Close']/Quarter_data['Preclose'] - 1
Quarter_data
```
